```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
from tensorflow import keras
from keras.models import load_model


# Load the trained model
model = load_model("/content/drive/MyDrive/htr/models/model2.h5")

# Load the input image and convert to grayscale
img = cv2.imread('/content/drive/MyDrive/htr/models/aa.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
# Threshold the image to binary
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

# Find contours of letters
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Sort the contours from left to right
contours = sorted(contours, key=lambda ctr: cv2.boundingRect(ctr)[0])

# Loop over the contours
for i, ctr in enumerate(contours):
    # Get the bounding box of the contour
    x, y, w, h = cv2.boundingRect(ctr)
    
    # Extract the letter image from the thresholded image
    letter_im = thresh[y:y+h, x:x+w]
    letter_img = cv2.bitwise_not(letter_im)
    cv2_imshow(letter_img)
    # Resize the letter image to 50x50
    letter_img = cv2.resize(letter_img, (50, 50), interpolation=cv2.INTER_AREA)
    print(cv2_imshow(letter_img))
    # Reshape the image to (1, 50, 50, 1)
    letter_img = letter_img.reshape((1, 50, 50, 1))
    
    # Normalize the image
    letter_img = letter_img / 255.0
    # print(letter_img)
    # Make a prediction using the model
    prediction = model.predict(letter_img)
    print(prediction)
    print(prediction.argmax(), len(prediction[0]))
    # Get the predicted letter
    dc = {1:'А', 2:'Б', 3:'В', 4:'Г', 5:'Д', 6:'Е', 7:'Ё', 8:'Ж', 9:'З', 10:'И', 11:'Й', 12:'К', 13:'Л', 14:'М', 15:'Н', 16:'Ң',17:'О',18:'Ө', 19:'П', 20:'Р', 21:'С', 22:'Т', 23:'У', 24:'Ү', 25:'Ф', 26:'Х', 27:'Ц', 28:'Ч', 29:'Ш', 30:'Щ', 31:'Ъ', 32:'Ы', 33:'Ь', 34:'Э', 35:'Ю', 36:'Я'}
    predicted_letter = dc[prediction.argmax()]
    print(predicted_letter)
    # Draw the predicted letter on the original image
    cv2.putText(img, predicted_letter, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

# Display the result
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
